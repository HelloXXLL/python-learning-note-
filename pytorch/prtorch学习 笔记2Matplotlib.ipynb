### 颜色
- b蓝色
- g绿色
- r red
- c 蓝绿色
- m 洋红色
-y yellow
- k 黑色
-w 白色
### 标记点用的符号
- o 圆形
- * 
- +
- x
#### 连接线条的形状
- -实线
- --虚线
- -.点实线
- :点线


```python
#群魔乱舞qwq
a=np.random.randn(30)
b=np.random.randn(30)
c=np.random.randn(30)
d=np.random.randn(30)
plt.plot(a,"r--o",b,"b-*",c,"g-.+",d,"m:x")
```

标签图例

子图

```python
a=np.random.randn(30)
b=np.random.randn(30)
c=np.random.randn(30)
d=np.random.randn(30)
fig=plt.figure()#fig定义一个实例，然后fig.add_subplot往实例里面放图
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
ax4=fig.add_subplot(2,2,4)
A,=ax1.plot(a,"r--o")
ax1.legend([A],["A"])
B,=ax2.plot(a,"b-*")
ax1.legend([B],["B"])
C,=ax3.plot(a,"g-.+")
ax1.legend([C],["C"])
D,=ax4.plot(a,"m:x")
ax1.legend([D],["D"])
```

散点图

```python
np.random.seed(42)
x=np.random.randn(30)
y=np.random.randn(30)
plt.scatter(x,y,c="g",marker="o",label="(X,Y)")
plt.title("Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(loc=1)
plt.show()
#loc=0图例使用最好的位置 loc=1强制右上角 loc=2强制左上角loc=3强制左下角loc=4强制右上角
```

直方图

```python
np.random.seed(42)
x=np.random.randn(1000)
plt.hist(x,bins=20,color="g")#bins直方图 条纹的数量
plt.title("Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

饼图

```python
labels=['Dos','Cats','Birds']
sizes=[15,50,35]#确定比例
plt.pie(sizes,explode=(0,0,0.1),labels=labels,autopct="%1.1f%%",startangle=90)#explode 确定数据之间的间隔 浮点数精度 绘制第一块饼图与x轴正方向呈的夹角大小
plt.axis('equal')#必须加上xy轴保持一致大概才能呈现圆形
plt.show()
```
